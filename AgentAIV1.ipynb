```python
from crewai import Agent

# Define a basic agent
general_agent = Agent(
    role='Movie  Agent',
    goal='Find Theaters to watch Movies requested by user with description',
    backstory='Need to find theaters where movies are screened.You can access information from https://in.bookmyshow.com/ and https://ticketnew.com/movies for schedule'
)
```

```python
import os
from crewai import Agent, Task, Crew, Process
from langchain.tools import tool
import requests
from bs4 import BeautifulSoup
import json

# 1. Define the Scraping Tool
@tool("TicketNew Scraper")
def scrape_ticketnew(url: str) -> str:
    """Scrapes movie data from a specific TicketNew URL."""
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Placeholder for actual TicketNew CSS selectors (e.g., class names for movie titles, cinemas)
    # This needs to be adapted to the specific structure of the TicketNew page
    movies = []
    for movie_card in soup.select('.movie-card-class'): # Update selector
        title = movie_card.select_one('.movie-title-class').text.strip() # Update selector
        movies.append({"title": title})

    return json.dumps(movies)

# 2. Setup LLM and Environment
# Ensure OPENAI_API_KEY is set in your environment
os.environ["OPENAI_MODEL_NAME"] = "gpt-4"

# 3. Define Agents
data_extractor = Agent(
    role='TicketDataExtractor',
    goal='Accurately extract movie titles and showtimes from TicketNew',
    backstory='An expert data researcher specialized in web scraping and structuring entertainment data.',
    tools=[scrape_ticketnew],
    verbose=True,
    allow_delegation=False
)

# 4. Define Tasks
extraction_task = Task(
    description='Extract all movie details from {url}',
    expected_output='A JSON list of movie titles and times.',
    agent=data_extractor
)

# 5. Form the Crew
crew = Crew(
    agents=[data_extractor],
    tasks=[extraction_task],
    process=Process.sequential
)

# 6. Execute
# Example URL - Replace with a real TicketNew city/movie URL
result = crew.kickoff(inputs={'url': 'https://www.ticketnew.com/movies/city'})
print(result)
```

```python

from crewai_tools import ScrapeElementFromWebsiteTool

# Define CSS selector for the links to child pages (e.g., in a menu or list)
# Example: 'a.child-page-link'
link_extractor = ScrapeElementFromWebsiteTool(
    website_url="https://ticketnew.com/movies/chennai",
    css_selector="a.child-page-link"
)


child_page_links = link_extractor.run()
results = []
for link in child_page_links:
    # Initialize tool for each child page
    child_scraper = ScrapeElementFromWebsiteTool(
        website_url=link,
        css_selector="https://ticketnew.com/movies/chennai/" # The specific element to scrape
    )
    content = child_scraper.run()
    results.append(content)
```
