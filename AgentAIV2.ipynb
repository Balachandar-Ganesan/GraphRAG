```python
from crewai import Agent

# Define a basic agent
general_agent = Agent(
    role='Movie  Agent',
    goal='Find Theaters to watch Movies requested by user with description',
    backstory='Need to find theaters where movies are screened.You can access information from https://in.bookmyshow.com/ and https://ticketnew.com/movies for schedule'
)
```

```python
from crewai import Task

# Define a task for the agent
planning_task = Task(
    description='Find Theaters Near Cognizant Siruseri Chennai office where Tamil Movies are running currently',
    expected_output='List of Theaters and address with Movie Name,Next earliest Date where movie can be watched',
    agent=general_agent
)
```

```python
from crewai import Crew, Process
from langchain_openai import ChatOpenAI
# Assemble the crew
crew = Crew(
    agents=[general_agent],
    tasks=[planning_task],
    process=Process.sequential,
    verbose=True
)
llm = ChatOpenAI(model="gpt-4-turbo-preview")
```

```python
result = crew.kickoff()

# Display the final output
print(result)
```

```python
import requests
from bs4 import BeautifulSoup
from crewai.tools import tool

@tool("element_scraper")
def element_scraper(url: str = "https://ticketnew.com/movies/chennai"):
    """
    Fetches a webpage and extracts all URLs from anchor tags.
    """
    try:
        # Fetch the HTML content
        response = requests.get(url)
        response.raise_for_status()  # Check for bad response

        # Parse the content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all anchor tags and extract the 'href' attribute
        urls = []
        for anchor in soup.find_all('a', href=True):
            href = anchor['href']
            # You might want to filter or resolve relative URLs here
            if href.startswith('/movies/'):
              href=href.replace("/movies/","https://ticketnew.com/movies/")
            urls.append(href)
        return urls

    except requests.exceptions.RequestException as e:
        print(f"Error fetching the URL: {e}")
        return []
```

```python
#print(element_scraper("https://ticketnew.com/movies/kolkata"))
```

```python
data_extractor = Agent(
    role='Data Extractor',
    goal='Extract movie names,theater name, theater address',
    backstory='Focused expert who can get movie name,theater name and address from a URL.',
    tools=[element_scraper] # Pass the tool function itself
)
```

```python
data_extraction_task = Task(
    description=(
        "1. Visit every URL available with the agent"
        "2. Get the list of theaters and address"
        "4. Share the information in bulletted list format "
    ),
    expected_output="Movie Name,Theater Address,Distance to Theater Address from Cognizant Siruseri office",
    agent=data_extractor,
)
```

```python
from crewai import Crew, Process

# (Code to define agents and tasks goes here)

my_crew = Crew(
    agents=[data_extractor],
    tasks=[data_extraction_task],
    process=Process.sequential,
)

# Kick off the crew with inputs
result = my_crew.kickoff()

# Print the result
print(result)
```
