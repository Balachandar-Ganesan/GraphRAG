```python
import requests
from bs4 import BeautifulSoup
from crewai.tools import tool

@tool("element_scraper")
def element_scraper(url: str = "https://ticketnew.com/movies/chennai"):
    """Scrapes movie links from TicketNew based on a given URL."""
    url = "https://ticketnew.com/movies/chennai"

    # Adding a User-Agent header helps avoid being blocked as a bot
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }

    try:
        # 1. Fetch the webpage content
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status() # Check for errors

        # 2. Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # 3. Find all anchor tags <a> with an 'href' attribute
        movie_links = []
        for link in soup.find_all('a', href=True):
            href = link['href']

            # 4. Filter for links containing "movie-detail"
            if "-movie-detail-" in href:
                # Convert relative URLs to absolute URLs if necessary
                full_url = href if href.startswith('http') else f"https://ticketnew.com{href}"
                movie_links.append(full_url)

        # Remove duplicates by converting to a set and back to a list
        return list(set(movie_links))

    except Exception as e:
        return f"Error: {e}"
```

```python
data_extractor = Agent(
    role='Data Extractor',
    goal='Extract movie names,theater name, theater address',
    backstory='Focused expert who can get movie name,theater name and address by visiting the URL.',
    tools=[element_scraper] # Pass the tool function itself
)
```

```python
data_extraction_task = Task(
    description=(
        "List All Movies with at least one Theater"

    ),
    expected_output="Movie Name,Theater Address,Distance to Theater  from Cognizant Siruseri office in Bulletted List",
    agent=data_extractor,
)
```

```python
from crewai import Crew, Process

# (Code to define agents and tasks goes here)

my_crew = Crew(
    agents=[data_extractor],
    tasks=[data_extraction_task],
    process=Process.sequential,
)

# Kick off the crew with inputs
result = my_crew.kickoff()

# Print the result
print(result)
```

```python
data_extractor = Agent(
    role='Local Cinema Analyst',
    goal='Extract movie details and identify theaters near Cognizant Siruseri',
    backstory='An expert in spatial data who filters entertainment options based on proximity to the Siruseri tech hub.',
    tools=[element_scraper], # Adding a search tool is highly recommended
    verbose=True
)

data_extraction_task = Task(
    description=(
        "1. Use element_scraper to get list of Individial URLS \n"
        "2. Get movie names and theater addresses from above URLS.\n"
        "3. For each address, determine the distance from Cognizant, SIPCOT IT Park, Siruseri.\n"

    ),
    expected_output="A bulleted list containing: Movie Name, Theater Name, Address, and Estimated Distance from Cognizant Siruseri.",
    agent=data_extractor
)
```

```python
from crewai import Crew, Process

# (Code to define agents and tasks goes here)

my_crew = Crew(
    agents=[data_extractor],
    tasks=[data_extraction_task],
    process=Process.sequential,
)

# Kick off the crew with inputs
result = my_crew.kickoff()

# Print the result
print(result)
```
