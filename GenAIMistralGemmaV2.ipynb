```python
from google.colab import userdata

# Access the secret value
MISTRAL_API_KEY = userdata.get('MISTRAL_API_KEY')
```

```python
from mistralai.client import Mistral
from mistralai.client.models import UserMessage, AssistantMessage
```

```python
def generate_code(strRequest):
 client = Mistral(api_key=MISTRAL_API_KEY)
 messages = [
    # Optional: You can start with a system message, though not strictly required
    # for simple code generation, it helps constrain the output.
    # SystemMessage(content="You are a helpful coding assistant that only responds with Python code."),

    UserMessage(
        content=strRequest
    )
]
 chat_response = client.chat.complete(
        model="codestral-latest",
        messages=messages,
    )

 generated_code = chat_response.choices[0].message.content
 print("Generated Code:")
 print(generated_code)
```

```python
generate_code("Generate code for XOrShift in Python")
```

```python
strContent ="""
I have a table in SQL.
It has columns Runs,Innings and MatchDate.
Write a query to get runs scored during First Innings every year
"""
generate_code(strContent)
```

```python
strContent ="""
Write a Code in Python for Email Validation
using regex and Decorators
"""
generate_code(strContent)
```
